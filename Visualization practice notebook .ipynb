## Matplotlib 

### Scatter plots

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(00,10,90)
y = np.sin(x)
```

```python
#Basic plot
plt.scatter(x,y, label = 'Sin (x)')
plt.title ('Plot of Sin')
plt.legend()
plt.show
```

```python
#add some x-tics
plt.figure(figsize=(10,6)) # specify size of plot
plt.scatter(x,y, label = 'Sin(x)')

plt.xlabel('X values')
plt.ylabel('Y values')

plt.title ('Plot of Sin')
plt.legend()
plt.show
```
